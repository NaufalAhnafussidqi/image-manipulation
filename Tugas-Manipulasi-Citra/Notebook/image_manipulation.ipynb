```python
img1 = cv.imread("image1.jpg")
img2 = cv.imread("image2.jpg")
```

```python
Jawaban Analisis Bagian B
1. Mengapa dua citra harus memiliki ukuran yang sama sebelum image blending?

Jawaban:

Karena proses image blending menggabungkan nilai pixel dari dua citra pada posisi yang sama. Jika ukurannya berbeda, setiap pixel tidak memiliki pasangan yang sesuai sehingga operasi tidak dapat dilakukan dengan benar.

2. Apa risiko resize terhadap kualitas citra?

Jawaban:

Resize dapat menyebabkan hilangnya detail, penurunan ketajaman, atau distorsi visual, terutama jika ukuran citra diperkecil atau diperbesar secara signifikan.

3. Apakah resize dapat mengubah proporsi objek?

Jawaban:

Ya. Jika rasio aspek (aspect ratio) tidak dipertahankan, objek pada citra dapat terlihat lebih lebar atau lebih tinggi dibanding bentuk aslinya.
```

```python
c = 255 / np.log10(1 + np.max(img1_resize))
```

```python
img_log = c * np.log10(img1_resize + 1)
img_log = np.array(img_log, dtype=np.uint8)
```

```python
Jawaban Analisis
1. Apa efek terhadap area gelap?

Jawaban:

Transformasi logaritmik meningkatkan intensitas pixel pada area gelap sehingga detail yang sebelumnya kurang terlihat menjadi lebih jelas.

2. Apa efek terhadap area terang?

Jawaban:

Area terang mengalami kompresi intensitas sehingga perubahan brightness tidak terlalu besar dibanding area gelap.

3. Mengapa sering digunakan untuk meningkatkan detail area gelap?

Jawaban:

Karena fungsi logaritma memberikan peningkatan yang lebih besar pada nilai pixel rendah dibanding nilai pixel tinggi.

4. Apa risikonya jika citra sudah sangat terang?

Jawaban:

Citra dapat kehilangan kontras dan beberapa detail pada area terang menjadi kurang jelas akibat kompresi intensitas yang berlebihan.
```

```python
img_normalized = img1_resize / 255.0
```

```python
def gamma_transform(image, gamma):
    return np.uint8((image ** gamma) * 255)
```

```python
gamma05 = gamma_transform(img_normalized, 0.5)
gamma1  = gamma_transform(img_normalized, 1)
gamma2  = gamma_transform(img_normalized, 2)
gamma25 = gamma_transform(img_normalized, 2.5)
```

```python
plt.figure(figsize=(20,5))

plt.subplot(1,5,1)
plt.imshow(img1_resize)
plt.title("Asli")
plt.axis("off")

plt.subplot(1,5,2)
plt.imshow(gamma05)
plt.title("Gamma=0.5")
plt.axis("off")

plt.subplot(1,5,3)
plt.imshow(gamma1)
plt.title("Gamma=1")
plt.axis("off")

plt.subplot(1,5,4)
plt.imshow(gamma2)
plt.title("Gamma=2")
plt.axis("off")

plt.subplot(1,5,5)
plt.imshow(gamma25)
plt.title("Gamma=2.5")
plt.axis("off")

plt.show()
```

```python
gammas = [gamma05, gamma1, gamma2, gamma25]
titles = ["0.5", "1", "2", "2.5"]

plt.figure(figsize=(16,8))

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.hist(gammas[i].ravel(), bins=256)
    plt.title("Gamma = " + titles[i])

plt.tight_layout()
plt.show()
```

```python
Jawaban Analisis
1. Apa yang terjadi ketika γ < 1?

Jawaban:

Intensitas pixel meningkat sehingga citra tampak lebih terang dan detail pada area gelap lebih mudah terlihat.

2. Apa yang terjadi ketika γ = 1?

Jawaban:

Citra tidak mengalami perubahan karena nilai pixel tetap sama.

3. Apa yang terjadi ketika γ > 1?

Jawaban:

Intensitas pixel menurun sehingga citra menjadi lebih gelap dan area terang menjadi lebih terkendali.

4. Bagaimana gamma digunakan untuk koreksi pencahayaan?

Jawaban:

Gamma correction dapat mencerahkan citra yang terlalu gelap (γ<1) atau mengurangi kecerahan citra yang terlalu terang (γ>1), sehingga tampilan menjadi lebih seimbang.

5. Kapan gamma lebih cocok dibanding logaritmik?

Jawaban:

Transformasi gamma lebih cocok ketika diperlukan pengaturan tingkat kecerahan secara fleksibel, sedangkan transformasi logaritmik lebih efektif untuk meningkatkan detail pada area gelap.
```
